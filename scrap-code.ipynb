```python
import requests
from bs4 import BeautifulSoup
import numpy as np

URL = "https://schpedia.ir/%D8%A7%D8%B3%D8%AA%D8%A7%D9%86"
page = requests.get(URL)

results = BeautifulSoup(page.content, "html.parser")
provinces = results.find_all("div", class_="card-deck col-lg-4 col-sm-6 my-3 m-0 px-0")

P_name, P_link= [], []
for p in provinces:
    P_name.append(p.find("h2", class_="mt-4 mb-2").text.strip())
    P_link.append('https://schpedia.ir' + p.find("h2", class_="mt-4 mb-2").find("a")['href'])
len(P_name), len(P_link)
```

```python
City = []
for l in P_link:
    C = []
    page = requests.get(l)
    results = BeautifulSoup(page.content, "html.parser")
    C.extend(results.find_all("div", class_="card-deck col-lg-4 col-sm-6 my-3 m-0 px-0"))
    C.extend(results.find_all("div", class_="col-lg-4 col-6 py-2 px-2"))
    City.append(C)
len(City)
```

```python
C_name, C_link = {}, {}
for i in range(31):
    C_name[P_name[i]] = []
    C_link[P_name[i]] = []
    for x in City[i]:
        try:
            C_name[P_name[i]].append(x.find("h2", class_="mb-0 p-3").text.strip())
            C_link[P_name[i]].append('https://schpedia.ir' + x.find("h2", class_="mb-0 p-3").find("a")['href'])
            
        except:
            C_name[P_name[i]].append(x.find("a").text.strip())
            C_link[P_name[i]].append('https://schpedia.ir' + x.find("a")['href'])
len(C_name), len(C_link)
```
